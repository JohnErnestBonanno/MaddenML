```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn import tree
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

%matplotlib inline
```

```python
df = pd.read_csv("player_stat_18.csv")
```

```python
#split positions into offense, defense, and special teams
off_pos_list = ['C','FB','HB','LG','LT','QB','RG','RT','TE','WR']
def_pos_list = ['CB','DT','FS','LE','LOLB','MLB','RE','ROLB','SS']
sTeams_pos_list = ['K','P']
```

```python
#Creates a new function to put each position into its own category.  
def define_pos(x):
    if x in off_pos_list:
        return('Off')
    if x in def_pos_list:
        return('Def')
    if x in sTeams_pos_list:
        return('sTeams')
    else:
        return("Missing")
```

```python
df['pos_class'] = df['Position'].apply(define_pos)
```

```python
extra_cols = ['Dev Trait','Last Name', 'First Name', 'Position','Birthdate','Years Pro','College','portraitId','Handedness']

#extra columns
df.drop(extra_cols, inplace=True, axis=1)
```

```python
le = preprocessing.LabelEncoder()
le.fit(df.pos_class)
```

```python
list(le.classes_)
```

```python
le.transform(df.pos_class)
```

```python
df.pos_class = le.transform(df.pos_class)
```

```python
y = df.pos_class
#create training and testing variables
x_train, x_test, y_train, y_test = train_test_split(df,y,test_size = 0.2)
clf = DecisionTreeClassifier()
clf.fit(x_train,y_train)
pred = clf.predict(x_test)
```

```python
acc = accuracy_score(pred, y_test)
print(acc)
```
